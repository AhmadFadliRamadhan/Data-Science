```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as BS

response = requests.get(url="https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(nominal)")
print(response.status_code)
```

```python
parse = BS(response.content, 'html.parser')

title = parse.find(id="firstHeading")
print(title.string)

soup = BS(response.text, 'html.parser')
table = parse.find('table', {'class': "wikitable"})
```

```python
df = pd.read_html(str(table))
df = pd.DataFrame(df[0][1:])
df.head()
```

```python
df.columns = df.columns.get_level_values(1)
df.head()
```
